```python
### testing code for spacing and other test((2019)
## function update initial code or marchig cubes is updated at source so newclassic methodis used (2020)
a = measure.marching_cubes_classic(vol, 0, spacing=(0.1, 0.1, 0.1))
```

```python
# Extract a 2D surface mesh from a 3D volume (F=0)
verts, faces = measure.marching_cubes_classic(vol, 0, spacing=(0.1, 0.1, 0.1))
# Create a 3D figure
love_fig = plt.figure(figsize=(24,16))
axis= love_fig.add_subplot(111, projection='3d')


axis.plot_trisurf(verts[:, 0], verts[:,1], faces, verts[:, 2],cmap='Spectral', lw=1)


axis.view_init(15, -15)

# ax.set_title(u"Made with ❤ (and Python)", fontsize=15) # if you have Python 3
axis.set_title(" O mea Nerine thymo mihi dulcior Hyblae,candidior cycnis,\n hedera formosior alba,cum primum pasti repetent praesepia tauri,\n si qua tui cunicularius habet te cura, uenito.", fontsize=20)
## translation in case anyone needs a proposal poem in latin  :
#O my Sea-Nymph sweeter to me than thyme of Hybla,
#whiter than the swan, lovelier than pale ivy,
#as soon as the pastured bulls seek the yard again,
#if you care at all for this engineer , come!
# Show me some love ^^
plt.show()
```
